```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../dataset/cleaned_NEM.csv')
```

```python
df = df.drop(columns=['Unnamed: 0'])
```

```python
# Check head
df.head()
```

```python
# Check tail
df.tail()
```

```python
# Calculate the statistical metrics for each relevant column
metrics = {
    "Mean": df[["Close", "Volume", "Open", "High", "Low", "Adj Close"]].mean(),
    "Median": df[["Close", "Volume", "Open", "High", "Low", "Adj Close"]].median(),
    "Mode": df[["Close", "Volume", "Open", "High", "Low", "Adj Close"]].mode().iloc[0],  # taking first mode in case of multiple
    "Range": df[["Close", "Volume", "Open", "High", "Low", "Adj Close"]].max() - df[["Close", "Volume", "Open", "High", "Low", "Adj Close"]].min(),
    "Variance": df[["Close", "Volume", "Open", "High", "Low", "Adj Close"]].var(),
    "Standard Deviation": df[["Close", "Volume", "Open", "High", "Low", "Adj Close"]].std()
}
```

```python
# Convert the dictionary to a DataFrame for display
metrics_df = pd.DataFrame(metrics)
```

```python
# Set display options to prevent line breaks in the output
pd.set_option('display.width', 1000)        # Set the display width to a large enough value
pd.set_option('display.max_columns', None)  # Ensure all columns are displayed


print(metrics_df)
```

# Histogram Chart

```python
# Plotting a histogram of gold prices over time using the "Price" column
plt.figure(figsize=(15, 6))
plt.hist(df["Close"], bins=50, edgecolor="black", alpha=0.6, color='grey')
plt.title("Histogram chart of gold price")
plt.xlabel("Gold Price")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Box plot

```python
plt.figure(figsize=(15, 6))

data = df['Close']

plt.boxplot(data, labels=['Box Plot of Gold Price'], patch_artist=True, boxprops=dict(facecolor='skyblue', color='blue'),)


plt.ylabel('Price')
```

# Scatter Plot

```python
# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])

# Plotting the scatter plot
plt.figure(figsize=(10, 6))
plt.scatter(df['Date'], df['Close'], s=5, alpha=0.6)  # Using small points for scatter plot effect
plt.title("Gold Prices Over Time (Scatter Plot)")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.grid(True)
plt.show()
```
