```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import f_oneway
```

```python
# Đọc dữ liệu từ các tệp
file_gc = pd.read_csv('../dataset/cleaned_gc=f.csv')
file_gld = pd.read_csv('../dataset/cleaned_GLD.csv')
```

```python
# Extract the 'Close' columns from both DataFrames
close_df1 = file_gc['Close']
close_df2 = file_gld['Close']

# Calculate the Pearson correlation coefficient between the two 'Close' columns
pearson_corr = close_df1.corr(close_df2)

pearson_corr
```

```python
# Scatter plot of the 'Close' prices to visualize their correlation

plt.scatter(close_df1, close_df2, alpha=0.5, color='green')
plt.xlabel('GC=F Close Prices')
plt.ylabel('GLD Close Prices')
plt.title('Scatter Plot of GC=F vs GLD Close Prices')

plt.grid(True)
plt.tight_layout()
plt.show()
```
