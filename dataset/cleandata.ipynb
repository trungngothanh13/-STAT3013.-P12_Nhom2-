3.2 Drop unnecessary column

3.3 Handling missing data

- Checking the missing data in each column

```python
df.isnull().sum()
```

```python
#Fill missing value 
df['Courier Status'].fillna('unknown', inplace=True)
df['promotion-ids'].fillna('no promotion', inplace=True)
```

```python
missing_cat_cols = [ 'ship-city', 'ship-state','ship-postal-code']
missing_num_cols = ['Amount']
```

```python
for col in missing_cat_cols:
    mode_value = df[col].mode()[0]
    df[col] = df[col].fillna(mode_value)
```

```python
for col in missing_num_cols:
    median_value = df[col].median()
    df[col] = df[col].fillna(median_value)
```

```python
df.isnull().sum()
```

- Missing Values in the Courier Status and promotion-ids has been Filled Successfully.
- Missing Values in the ship-city and ship-state and ship-postal-code has been Filled Successfully. using mode of each column

3.4 Rename Columns

```python
mapper = {'Order ID':'order_ID', 'Date':'date', 'Status':'ship_status','Fulfilment':'fullfilment',
          'ship-service-level':'service_level', 'Style':'style', 'SKU':'sku', 'Category':'product_category', 
          'Size':'size', 'ASIN':'asin', 'Courier Status':'courier_ship_status', 'Qty':'order_quantity', 
          'Amount':'order_amount_($)', 'ship-city':'city', 'ship-state':'state', 'ship-postal-code':'zip', 
          'promotion-ids':'promotion','B2B':'customer_type'}
df.rename(columns=mapper, inplace =True)
```

- Convert Amount column values to Dollar based on exchange rate

```python
# Convert INR to USD using an exchange rate of 1 INR = 0.012 USD
exchange_rate = 0.0120988
df['order_amount_($)'] = df['order_amount_($)'].apply(lambda x: x * exchange_rate)

```
