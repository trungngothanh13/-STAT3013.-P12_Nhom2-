```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('dataset/cleaned_gold.csv')
df.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
# Check head
df.head()
```

```python
# Check tail
df.tail()
```

```python
# Calculate the statistical metrics for each relevant column
metrics = {
    "Mean": df[["Price", "Volume", "Open", "High", "Low"]].mean(),
    "Median": df[["Price", "Volume", "Open", "High", "Low"]].median(),
    "Mode": df[["Price", "Volume", "Open", "High", "Low"]].mode().iloc[0],  # taking first mode in case of multiple
    "Range": df[["Price", "Volume", "Open", "High", "Low"]].max() - df[["Price", "Volume", "Open", "High", "Low"]].min(),
    "Variance": df[["Price", "Volume", "Open", "High", "Low"]].var(),
    "Standard Deviation": df[["Price", "Volume", "Open", "High", "Low"]].std()
}
```

```python
# Convert the dictionary to a DataFrame for display
metrics_df = pd.DataFrame(metrics)
```

```python
# Set display options to prevent line breaks in the output
pd.set_option('display.width', 1000)        # Set the display width to a large enough value
pd.set_option('display.max_columns', None)  # Ensure all columns are displayed


print(metrics_df)
```

# Histogram Chart

```python
# Plotting a histogram of gold prices over time using the "Price" column
plt.figure(figsize=(15, 6))
plt.hist(df["Price"], bins=50, edgecolor="black", alpha=0.6, color='grey')
plt.title("Histogram chart of gold price")
plt.xlabel("Gold Price")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

# Box plot

```python
plt.figure(figsize=(15, 6))

data = df['Price']

plt.boxplot(data, labels=['Box Plot of Gold Price'], patch_artist=True, boxprops=dict(facecolor='skyblue', color='blue'),)


plt.ylabel('Price')
```

# Scatter Plot

```python
# Plotting the scatter plot for gold prices over time
plt.figure(figsize=(10, 5))
plt.scatter(df['Date'], df['Price'], color='red', s=10)
plt.title('Gold Prices Over Time (Scatter Plot)')
plt.xlabel('Date')
plt.ylabel('Price')
plt.grid(True)
plt.show()
```
