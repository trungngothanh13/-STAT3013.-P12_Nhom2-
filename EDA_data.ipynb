# 1. Statistics Description: 

1.1 Calculate the central trend measures and Determination of dispersion measures

```python
# Convert relevant columns to numeric types for calculation
df['order_amount_($)'] = pd.to_numeric(df['order_amount_($)'], errors='coerce')
```

```python
# Group by product category
category_grouped = df.groupby('product_category')
```

```python
# Calculate count, mean, median, mode, range, variance, and standard deviation for order_amount_($)
count_category = category_grouped['order_amount_($)'].count()
mean_amount = category_grouped['order_amount_($)'].mean()
median_amount = category_grouped['order_amount_($)'].median()
mode_amount = category_grouped['order_amount_($)'].agg(lambda x: x.mode().iloc[0] if not x.mode().empty else None)
range_amount = category_grouped['order_amount_($)'].agg(lambda x: x.max() - x.min())
variance_amount = category_grouped['order_amount_($)'].var()
std_dev_amount = category_grouped['order_amount_($)'].std()
sum_amount = category_grouped['order_amount_($)'].sum()
```

```python
# Combine all measures into a DataFrame
dispersion_df = pd.DataFrame({
    'Count': count_category,
    'Mean Amount ($)': mean_amount,
    'Median Amount ($)': median_amount,
    'Mode Amount ($)': mode_amount,
    'Range Amount ($)': range_amount,
    'Variance Amount ($)': variance_amount,
    'Standard Deviation ($)': std_dev_amount,
    'Sum ($)': sum_amount
})
```
